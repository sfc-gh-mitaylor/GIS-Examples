# 2. Ingest .grib File (from ERA)

File contains hourly temperatures at 2m from a selection of coordinates approximating London

## 2.a Read the file locally and push to a pandas dataframe (or direct to snowpark, not much difference for single node)

```python
#grib_loc = '/Users/mitaylor/Documents/GitHub/Untitled Folder/Untitled Folder/data/adaptor.mars.internal-1680538767.2985864-32006-16-d1ae6a24-8d91-4aff-9ef5-f5777509e10a.grib'
data=[]
data_lon = []
data_lat = []
data_time = []
data_temp = []

with open(grib_loc, 'rb') as stream:
    for i, msg in enumerate(pupygrib.read(stream), 1):
        lons, lats = msg.get_coordinates()
        time = msg.get_time()
        values = msg.get_values()
        data_lon.append(lons)
        data_lat.append(lats)
        data_time.append(time)
        data_temp.append(values)

cells = np.ravel(data_temp[0]).shape[0]
data_clean_df = pd.DataFrame()
data_clean_df['DATE'] = list(np.ravel([[i]*cells for i in data_time]))
data_clean_df['TIME'] = data_clean_df['DATE'].dt.time
data_clean_df['DATE'] = data_clean_df['DATE'].dt.date
data_clean_df['LONG'] = list(np.ravel(data_lon))
data_clean_df['LAT'] = list(np.ravel(data_lat))
data_clean_df['TEMP'] = list(np.ravel(data_temp))
data_clean_df.head()
        
```

```python
session.write_pandas(df=data_clean_df, table_name="GRIB_DATA_RAW_LOCAL", overwrite=True, auto_create_table=True)
```

## 2.b Or via a Sproc inside Snowflake

```python
@sproc(name='PUPYGRIB_READ', 
       packages=['snowflake-snowpark-python'], 
       imports=pupygrib.__path__,
       is_permanent=True, 
       replace=True,
       stage_location='@PIPELINE', 
       session=session)
def read_grib(session: Session, filename: str, target_table: str) -> str:
    import io
    import pupygrib
    import numpy as np
    file_data = io.BytesIO(session.file.get_stream(filename).read())
    
    data=[]
    data_lon = []
    data_lat = []
    data_time = []
    data_temp = []

    for i, msg in enumerate(pupygrib.read(file_data), 1):
        lons, lats = msg.get_coordinates()
        time = msg.get_time()
        values = msg.get_values()
        data_lon.append(lons)
        data_lat.append(lats)
        data_time.append(time)
        data_temp.append(values)

    cells = np.ravel(data_temp[0]).shape[0]
    data_clean_df = pd.DataFrame()
    data_clean_df['DATE'] = list(np.ravel([[i]*cells for i in data_time]))
    data_clean_df['TIME'] = data_clean_df['DATE'].dt.time
    data_clean_df['DATE'] = data_clean_df['DATE'].dt.date
    data_clean_df['LONG'] = list(np.ravel(data_lon))
    data_clean_df['LAT'] = list(np.ravel(data_lat))
    data_clean_df['TEMP'] = list(np.ravel(data_temp))

    session.write_pandas(df=data_clean_df, table_name=target_table, overwrite=True, auto_create_table=True)
    return f'Successfully written data to table {target_table}'
```

```python
# Execute Stored Procedure
filename = f"@{session.sql('LS @PIPELINE').collect()[1][0]}"
target_table = 'GRIB_DATA_RAW_SF'

print('filename:', filename)
print('target-table:', target_table)

```

```python
read_grib( filename, target_table)
```

# 3. Update Snowflake Table with Geospatial properties

## 3.1 Create Coordinates

```python
snowpark_df_gis = session.sql('''SELECT DATE, TIME, LONG, LAT, TEMP, ST_MAKEPOINT(o.LAT, o.LONG) as COORD
                               FROM GRIB_DATA_RAW_SF o''')
snowpark_df_gis.write.mode("overwrite").save_as_table("GRIB_DATA_GIS")
```

```python
snowpark_df_gis.limit(5).to_pandas()
```

## 3.2 Select Coordinates Within Polygon

```python
snowpark_df_gis_test = session.sql('''
SELECT o.date, o.time, o.long, o.lat, o.temp, ST_CONTAINS(nl_poly, o.coord) as in_n_london
FROM
GRIB_DATA_GIS o,
(SELECT(TO_GEOGRAPHY('POLYGON((51.5 0, 52 0.5, 51.5 1, 51.5 0))')) AS nl_poly)''')
```

```python
snowpark_df_gis_test.limit(5).to_pandas()
```

```python
snowpark_df_gis_test = snowpark_df_gis_test.filter((snowpark_df_gis_test["IN_N_LONDON"] == True))
snowpark_df_gis_test.to_pandas()
```

# 4. Query the data (Calculate Mean, grouped by Date, for all coordinates)

```python
df_viz = snowpark_df_gis_test.group_by("DATE").function("avg")("TEMP").to_pandas()
```

```python
import seaborn as sns

sns.lineplot(y="AVG(TEMP)", x="DATE", data=df_viz)
```
